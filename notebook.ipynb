### Hard Constraints
- Nobody rostered on a day that they cannot make it

### Soft Constraints
- Nobody rostered 4 or more times
- Nobody rostered back to back

### Optimise
- Minimise variance in total scores across days
- Maximise mean of total scores across days

```python
import pandas as pd
import numpy as np
from utility import *
from random import sample

INDIVIDUAL_DATE_SAMPLE = 20
```

```python
input_df = pd.read_csv('taboo_dates.csv')
dates = list(input_df.columns)[1:]

# input_df = input_df[['name', 'd1', 'd2', 'd3','d4']]
# dates = ['d1', 'd2', 'd3','d4']
# input_df.loc[4, 'd1'] = False
# input_df.loc[6, 'd1'] = False

print('input df:')
display(input_df.head())
```

```python
date_combinations = []
for date in dates:
    free_people = set(input_df['name'][input_df[date]].values)
    date_combination = get_perms_single_day(free_people)
    date_combinations.append(sample(date_combination, INDIVIDUAL_DATE_SAMPLE))
```

```python
%%time
all_combinations = get_comb_multi_day(date_combinations)
```

```python
%%time
all_combinations = get_comb_multi_day_b2b_constraint(date_combinations)
```

```python
print(f'Number of combinations: {len(all_combinations)}')
for combination in all_combinations[:3]:
    display(pd.DataFrame(combination, index=dates).drop(columns = ['people']))
```
