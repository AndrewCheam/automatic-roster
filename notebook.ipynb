```python
input_df = pd.read_csv('taboo_dates.csv')
dates = list(input_df.columns)[1:]

input_df = input_df[['name', 'd1']]
dates = ['d1']
input_df[[['d1']]] = True

print('input df:')
display(input_df)
```

```python
date_combinations = []
for date in dates:
    free_people = set(input_df['name'][input_df[date]].values)
    date_combination = get_perms(free_people, 6)
    date_combinations.append(date_combination)
```
