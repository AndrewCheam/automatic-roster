```python
from itertools import product

def get_all_combinations(date_combinations):
    combinations = list(product(*date_combinations))
    # Check the total number of combinations
    return combinations
```

## First lets make some fake data

In this example there are 3 dates: {a,b,c} and there are 3 roles: {ws_lead, guitarist, pianist}

columns: [name, a_free, b_free, c_free, ws_lead_prof, guiarist_prof, pianist_prof]

Name: string # Name of person. Is key
Dates: bool # Free or not free for that date
Roles: int # 0 if not proficient and range between 1 - 5 for proficiency

```python
dates = ['a', 'b', 'c']
input_col_names = ['name', 'a', 'b', 'c', 'ws_lead', 'guitarist', 'pianist']
input_data = [['Andrew', True, True, True, 3, 3, 0], ['Rissa', True, True, True, 3, 3, 0], ['Eliza', True, True, True, 0, 0, 2]]
```

```python
input_df = pd.DataFrame(input_data, columns=input_col_names)
print('input df:')
display(input_df)
```

```python
%%timeit
date_combinations = []
for date in dates:
    free_people = set(input_df['name'][input_df[date]].values)
    date_combination = get_perms(free_people, 3)
    date_combinations.append(date_combination)
all_combinations = get_all_combinations(date_combinations)

```

```python
for combination in all_combinations:
    display(pd.DataFrame(combination))
```

```python
ws_lead_free = list(input_df['name'][(input_df['a'] & input_df['ws_lead'] > 0)].values)
guitarist_free = list(input_df['name'][(input_df['a'] & input_df['guitarist'] > 0)].values)
pianist_free = list(input_df['name'][(input_df['a']) & (input_df['pianist'] > 0)].values)
```
