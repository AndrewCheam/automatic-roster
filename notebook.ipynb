### Hard Constraints
- Nobody rostered on a day that they cannot make it
- Nobody rostered more than their max roster limit
- Nobody rostered 4 or more times

### Soft Constraints
- Nobody rostered back to back (Currently still hard), being used as backtracking mechanism

### Optimise (through sorting only)
- Minimise variance in total scores across days
- Maximise mean of total scores across days

```python
date_combinations = []
for date in dates:
    free_people = set(input_df['name'][input_df[date]].values)
    date_combination = get_perms_single_day(free_people)
    print(f'Number of combinations for {date} is {len(date_combination)}')
    date_combinations.append(sample(date_combination, min(INDIVIDUAL_DATE_SAMPLE, len(date_combination))))
```

```python
%%time
all_combinations = get_comb_multi_day_b2b_constraint(date_combinations)
all_combinations_with_score = get_quarter_comb_with_score(all_combinations)
```
