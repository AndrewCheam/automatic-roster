```python
import pandas as pd
```

## Create availability_df

```python
df = pd.read_excel('temp_data.xlsx')
date_cols = df.columns[1:]
df[date_cols] = df[date_cols].notnull()
df = df.loc[~df['Names'].isin(['BENEDICT', 'MAVIS', 'ROBIN'])]
df = df.drop_duplicates().reset_index(drop=True)
df.columns = ['Names']+[str(x.date()) for x in date_cols]
display(df)
df.to_excel('q2_date_availability.xlsx')
```

## Create skill_matching data

```python
all_names = list(set(df['Names']))
cols = [
    'SOUND #1 (M)',
    'PROJECTIONIST (M)',
    'COORDINATOR (B)',
    'LIVESTREAM #1 (B)',
    'SOUND #1 (B)',
    'SOUND #2 (B)',
    'PROJECTIONIST (B)'
    ]
df = pd.DataFrame(columns=cols, index=all_names, data=False)


def update_skill_df(df:pd.DataFrame, role:str, members:list)->None:
    df.loc[members, role] = True
    return df

df = update_skill_df(df, 'SOUND #1 (M)', ['SINGYE', 'WONSEN', 'SHERMAN'])
df = update_skill_df(df, 'PROJECTIONIST (M)', ['CHEE FS', 'FRED KWAN', 'IAN MING', 'LAWRENCE', 'PAMELA CHEONG', 'HONGTO', 'STEVEN', 'ROBIN'])
df = update_skill_df(df, 'COORDINATOR (B)', ['WEI XUAN', 'YU XUAN', 'ANDREW', 'FAM', 'TITUS', 'JURINA', 'BENEDICT'])
df = update_skill_df(df, 'LIVESTREAM #1 (B)', ['SIAW ENG', 'HING GEE', 'SWEE GEOK', 'MU TIAN', 'TIMOTHY', 'YU TANG', 'ZHI QIAN'])
df = update_skill_df(df, 'SOUND #1 (B)', ['JAIRUS', 'JEFFERY', 'SHAWN', 'JEREMY', 'MATHIAS', 'TIMOTHY'])
df = update_skill_df(df, 'SOUND #2 (B)', ['JAIRUS', 'JEFFERY', 'SHAWN', 'JEREMY', 'MATHIAS', 'TIMOTHY'])
df = update_skill_df(df, 'PROJECTIONIST (B)', ['BENEDICT', 'ELLY', 'JURINA', 'MAVIS', 'SIAW ENG', 'TIMOTHY'])

df = df.sort_index()

df = df.drop(index='BENEDICT')
df = df.drop(index='MAVIS')
df = df.drop(index='ROBIN')
df = df.reset_index(names=['Names'])
display(df)
df.to_excel('skills_mapping.xlsx')
```
